Introduction

This project aims to determine the “best neighborhood” in Pittsburgh using a data-driven approach. Since the idea of “best” can vary depending on personal preferences, our group defined different metrics to evaluate neighborhoods.
For my part of the project, I focused on environmental quality by analyzing access to green spaces such as parks and recreational areas. Green spaces are important because they provide opportunities for exercise, relaxation, and social interaction, which contribute to overall quality of life.
To measure this, I used a dataset from the Western Pennsylvania Regional Data Center (WPRDC) and analyzed how many green spaces exist in each neighborhood.

The Metric:
The metric I used to evaluate neighborhoods is:
👉 Number of green spaces per neighborhood

This metric is based on the idea that neighborhoods with more parks and green areas offer better access to outdoor activities and healthier living environments.

Why this metric matters:
Encourages physical activity
Supports mental well-being
Provides community gathering spaces
Dataset Explanation

The dataset used in this analysis comes from the Western Pennsylvania Regional Data Center (WPRDC).
It contains information about green spaces across Pittsburgh
Each row represents a park or green space
It includes the neighborhood where each green space is located

This dataset allows us to count how many green spaces exist in each neighborhood.

```python
import pandas as pd   ## Import required libraries

import matplotlib.pyplot as plt
df = pd.read_csv("green_spaces.csv")  # Load the dataset from the project folder
df.head()
```

```python
green_counts = df['neighborhood'].value_counts() #Count green spaces per neighborhood
green_counts
```

```python
green_counts.head(10) #show top neiborhood
```

```python
green_counts.head(10).plot(kind='bar') # Create a bar chart of the top 10 neighborhoods
plt.title("Top Pittsburgh Neighborhoods by Green Spaces")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Green Spaces")
plt.show()
```

```python
best_neighborhood = green_counts.idxmax() # Find the neighborhood with the highest number of green spaces
best_value = green_counts.max()

print("Best Neighborhood:", best_neighborhood)
print("Number of Green Spaces:", best_value)
```

The analysis shows that:

East Liberty has the highest number of green spaces (12)
Other top neighborhoods include:
Central Business District
Beechview
South Side Slopes

The bar chart clearly shows that East Liberty stands out compared to other neighborhoods.

Conclusion

Based on this analysis, East Liberty is the best neighborhood by this metric because it has the most green spaces.

This result is important because green spaces are strongly linked to the quality of life. They encourage physical activity, reduce stress, and provide areas for community interaction. Neighborhoods with more green spaces tend to offer healthier and more enjoyable living environments for residents of all ages, including children, families, and older adults.

In addition, access to parks and outdoor areas can also improve social connections within a community. People are more likely to gather, exercise, and participate in outdoor activities when these spaces are available and accessible.

However, this analysis only considers one factor. Other important aspects such as safety, cost of living, education quality, transportation access, and job opportunities, could also influence what makes a neighborhood the “best.” Because of this, the results should be understood as one perspective rather than a complete definition of neighborhood quality.

Overall, this project shows how data can support a clear, evidence-based argument, but also highlights that real-world decisions are usually based on multiple factors, not just a single metric.
