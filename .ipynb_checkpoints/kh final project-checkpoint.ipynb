---
## Step 1: Import Libraries

```python
# pandas is a library that helps us work with data in tables (like Excel)
import pandas as pd

# matplotlib is a library that helps us create charts and graphs
import matplotlib.pyplot as plt

# numpy is a library for doing math operations
import numpy as np

# This line makes our charts look nicer with a clean white grid background
plt.style.use('seaborn-v0_8-whitegrid')

# This sets the default size for all our charts (12 inches wide, 8 inches tall)
plt.rcParams['figure.figsize'] = (12, 8)
```

---
## Step 2: Load the Data

We have 8 CSV files, one for each racial/ethnic category. We'll load them all and combine them to get total population counts.

```python
# This is a list of all our CSV file names
# A list in Python is made with square brackets []
# Each item in the list is a file name as a string (text in quotes)
# UPDATE THESE PATHS to match where your files are located!
files = [
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-white-alone.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-black-or-afric.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-asian-alone.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-hispanic-or-la.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-american-india.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-native-hawaiia.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-some-other-rac.csv',
    'school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over-two-or-more-ra.csv'
]

# pd.read_csv() reads a CSV file and turns it into a DataFrame (a table)
# files[0] gets the first item from our list (Python counts from 0, not 1)
# We're loading just the first file to see what the data looks like
df_sample = pd.read_csv(files[0])

# .shape tells us how many rows and columns are in our table
# The f before the quotes lets us put variables inside {} to print them
print(f"Shape: {df_sample.shape}")

# .columns gives us all the column names
# .tolist() converts it to a regular Python list so it prints nicely
print(f"\nColumns: {df_sample.columns.tolist()}")
```

```python
# .head() shows us the first 5 rows of our table
# This helps us see what the actual data looks like
df_sample.head()
```

---
## Step 3: Combine All Racial/Ethnic Groups

To get the total higher education rate, we need to sum up the populations across all racial groups for each neighborhood.

```python
# Create an empty list to store all our DataFrames
# We'll fill this list with data from each CSV file
all_dfs = []

# This is a "for loop" - it repeats the code inside for each item in our list
# Each time through the loop, "file" becomes the next file name from "files"
for file in files:
    
    # Read the current CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # .append() adds the DataFrame to our list
    all_dfs.append(df)

# len() counts how many items are in a list
# This should print 8 since we have 8 files
print(f"Loaded {len(all_dfs)} files")
```

```python
# Let's find the exact column names we need
# .columns.tolist() gets all column names as a list
cols = df_sample.columns.tolist()

print("Looking for relevant columns...")

# Loop through each column name
for col in cols:
    
    # .lower() converts text to lowercase so we can search more easily
    # "in" checks if one string is inside another string
    # This finds columns with 'college', 'graduate', or 'total' in the name
    if 'college' in col.lower() or 'graduate' in col.lower() or 'total' in col.lower():
        
        # Print the column name with some indentation for readability
        print(f"  - {col}")
```

```python
# Create a new empty DataFrame to store our combined results
# pd.DataFrame() creates an empty table
combined = pd.DataFrame()

# Copy the Neighborhood column from the first DataFrame
# all_dfs[0] is the first DataFrame in our list
# ['Neighborhood'] gets just that one column
combined['Neighborhood'] = all_dfs[0]['Neighborhood']

# Create new columns and set all values to 0
# We'll add to these as we loop through each racial group
combined['Total_Population'] = 0
combined['College_Undergraduate'] = 0
combined['Graduate_Professional'] = 0

# Loop through each DataFrame (one for each racial group)
for df in all_dfs:
    
    # These are the exact column names from the CSV files
    # We save them as variables to make the code easier to read
    total_col = 'Estimate; Total:'
    college_col = 'Estimate; Enrolled in school: - Enrolled in college, undergraduate years'
    grad_col = 'Estimate; Enrolled in school: - Graduate or professional school'
    
    # pd.to_numeric() converts text to numbers
    # errors='coerce' means if something can't be converted, make it NaN (missing)
    # .fillna(0) replaces any NaN values with 0
    # += means "add this to what's already there"
    
    # Add total population from this racial group to our running total
    combined['Total_Population'] += pd.to_numeric(df[total_col], errors='coerce').fillna(0)
    
    # Add college enrollment from this racial group to our running total
    combined['College_Undergraduate'] += pd.to_numeric(df[college_col], errors='coerce').fillna(0)
    
    # Add graduate school enrollment from this racial group to our running total
    combined['Graduate_Professional'] += pd.to_numeric(df[grad_col], errors='coerce').fillna(0)

# Show the first 10 rows of our combined data
combined.head(10)
```

---
## Step 4: Calculate Higher Education Rate

```python
# Add together college and graduate school to get total higher ed enrollment
# This creates a new column called 'Higher_Ed_Total'
combined['Higher_Ed_Total'] = combined['College_Undergraduate'] + combined['Graduate_Professional']

# Calculate the percentage: (higher ed students / total population) * 100
# This tells us what percent of the neighborhood is in higher education
combined['Higher_Ed_Rate'] = (combined['Higher_Ed_Total'] / combined['Total_Population']) * 100

# If a neighborhood has 0 population, we'd be dividing by 0, which gives infinity (inf)
# .replace() swaps out infinity values with NaN (Not a Number / missing)
# np.inf is positive infinity, -np.inf is negative infinity
# np.nan is how pandas represents missing data
combined['Higher_Ed_Rate'] = combined['Higher_Ed_Rate'].replace([np.inf, -np.inf], np.nan)

# Show the first 10 rows to check our work
combined.head(10)
```

```python
# Print some basic statistics about our higher education rate
print("Higher Education Rate Statistics:")

# .mean() calculates the average
# :.2f means format as a decimal with 2 places after the decimal point
print(f"  Mean: {combined['Higher_Ed_Rate'].mean():.2f}%")

# .median() finds the middle value when all values are sorted
print(f"  Median: {combined['Higher_Ed_Rate'].median():.2f}%")

# .min() finds the smallest value
print(f"  Min: {combined['Higher_Ed_Rate'].min():.2f}%")

# .max() finds the largest value
print(f"  Max: {combined['Higher_Ed_Rate'].max():.2f}%")

# .std() calculates standard deviation (how spread out the values are)
print(f"  Std Dev: {combined['Higher_Ed_Rate'].std():.2f}%")
```

---
## Step 5: Rank the Neighborhoods

```python
# .sort_values() sorts the DataFrame by a column
# 'Higher_Ed_Rate' is the column we want to sort by
# ascending=False means highest values first (descending order)
# .reset_index(drop=True) gives us fresh row numbers starting from 0
ranked = combined.sort_values('Higher_Ed_Rate', ascending=False).reset_index(drop=True)

# Create a new column called 'Rank' with numbers 1, 2, 3, etc.
# range(1, len(ranked) + 1) creates numbers from 1 to the number of neighborhoods
# len(ranked) gives us how many rows are in our DataFrame
ranked['Rank'] = range(1, len(ranked) + 1)

# Print a header (the emoji is just for fun!)
print("🏆 TOP 10 NEIGHBORHOODS BY HIGHER EDUCATION RATE:\n")

# Select only certain columns to display using double brackets [[ ]]
# .head(10) shows only the first 10 rows
top_10 = ranked[['Rank', 'Neighborhood', 'Higher_Ed_Rate', 'Higher_Ed_Total', 'Total_Population']].head(10)

# Display the table
top_10
```

```python
# Same thing but for the bottom 10 (worst) neighborhoods
print("📉 BOTTOM 10 NEIGHBORHOODS BY HIGHER EDUCATION RATE:\n")

# .tail(10) shows the last 10 rows instead of the first 10
bottom_10 = ranked[['Rank', 'Neighborhood', 'Higher_Ed_Rate', 'Higher_Ed_Total', 'Total_Population']].tail(10)

# Display the table
bottom_10
```

---
## Step 6: Visualize the Results

```python
# Get the top 15 neighborhoods for our chart
top_15 = ranked.head(15)

# Create a new figure (chart) with a specific size
# figsize=(12, 8) means 12 inches wide and 8 inches tall
plt.figure(figsize=(12, 8))

# Create a horizontal bar chart
# plt.barh() makes horizontal bars (h = horizontal)
# First argument: what goes on the y-axis (neighborhood names)
# Second argument: the length of each bar (higher ed rate)
# color='steelblue' sets the bar color
bars = plt.barh(top_15['Neighborhood'], top_15['Higher_Ed_Rate'], color='steelblue')

# Label the x-axis
# fontsize=12 makes the text size 12 points
plt.xlabel('Higher Education Rate (%)', fontsize=12)

# Label the y-axis
plt.ylabel('Neighborhood', fontsize=12)

# Add a title at the top
# fontweight='bold' makes the title bold
plt.title('Top 15 Pittsburgh Neighborhoods by Higher Education Rate', fontsize=14, fontweight='bold')

# Flip the y-axis so the highest ranked neighborhood is at the top
# By default, bar charts put the first item at the bottom
plt.gca().invert_yaxis()

# Add the actual percentage numbers at the end of each bar
# zip() pairs up each bar with its corresponding value
for bar, val in zip(bars, top_15['Higher_Ed_Rate']):
    
    # plt.text() adds text to the chart
    # val + 0.5 positions it slightly to the right of the bar end
    # bar.get_y() + bar.get_height()/2 centers it vertically on the bar
    # f'{val:.1f}%' formats the number with 1 decimal place and adds %
    # va='center' vertically aligns the text to the center
    plt.text(val + 0.5, bar.get_y() + bar.get_height()/2, f'{val:.1f}%', 
             va='center', fontsize=10)

# Adjust spacing so everything fits nicely
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Create a histogram to see the distribution of higher education rates
# A histogram shows how many neighborhoods fall into each range

# Create a new figure
plt.figure(figsize=(10, 6))

# Create the histogram
# .dropna() removes any missing values first
# bins=20 divides the data into 20 groups
# edgecolor='white' puts white borders around each bar
plt.hist(combined['Higher_Ed_Rate'].dropna(), bins=20, color='steelblue', edgecolor='white')

# Label the axes
plt.xlabel('Higher Education Rate (%)', fontsize=12)
plt.ylabel('Number of Neighborhoods', fontsize=12)

# Add a title
plt.title('Distribution of Higher Education Rates Across Pittsburgh Neighborhoods', fontsize=14, fontweight='bold')

# Add a vertical line showing the average (mean)
# plt.axvline() draws a vertical line
# color='red' makes it red
# linestyle='--' makes it dashed
# label= adds text for the legend
plt.axvline(combined['Higher_Ed_Rate'].mean(), color='red', linestyle='--', label=f"Mean: {combined['Higher_Ed_Rate'].mean():.1f}%")

# Show the legend (explains what the red line means)
plt.legend()

# Adjust spacing
plt.tight_layout()

# Display the chart
plt.show()
```

Save the ranked data so your teammates can combine it with their sub-metrics.

```python
# Create a clean DataFrame with only the columns we need to share
# .copy() makes a separate copy so we don't accidentally change the original
export_df = ranked[['Neighborhood', 'Higher_Ed_Rate', 'Rank']].copy()

# Rename the columns to be clearer
# This replaces all column names at once
export_df.columns = ['Neighborhood', 'Higher_Ed_Rate_Pct', 'Higher_Ed_Rank']

# Save the DataFrame to a CSV file
# .to_csv() writes the data to a file
# index=False means don't include the row numbers in the file
export_df.to_csv('higher_education_metric.csv', index=False)

# Print confirmation messages
print("✅ Saved to 'higher_education_metric.csv'")
print("\nYour teammates can merge this with their metrics using the 'Neighborhood' column.")
```
